```python
# Importing the data set
import pandas as pd

df = pd.read_csv("data.csv")
df.info()
```

```python
# Dropping 2 columns because they are not related to output
df = df.drop("Unnamed: 32", axis=1)
df = df.drop("id", axis=1)
df.info()
```

```python
#Viewing the statistical analysis
df.describe()
```

```python
# Create histograms to visualize data
import matplotlib.pyplot as plt

attributes = ['radius_mean', 'texture_mean', 'perimeter_mean',
                'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
                'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
                'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
                'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
                'fractal_dimension_se', 'radius_worst', 'texture_worst',
                'perimeter_worst', 'area_worst', 'smoothness_worst',
                'compactness_worst', 'concavity_worst', 'concave points_worst',
                'symmetry_worst', 'fractal_dimension_worst']

df[attributes].hist(bins=30, figsize=(20, 15), color='steelblue', edgecolor='black')

plt.suptitle("Data Visualization", fontsize=28, weight='bold')
plt.show()
```

```python
# Splitting for train and test
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Splitting the output we need to predict
df_train = train_set.drop("diagnosis", axis=1)
df_train_labels = train_set["diagnosis"].copy()

df_test = test_set.drop("diagnosis", axis=1)
df_test_labels = test_set["diagnosis"].copy()
```

```python
# Encode categorical class labels into numeric form using LabelEncoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_train_labels_prepared = label_encoder.fit_transform(df_train_labels)
df_test_labels_prepared = label_encoder.transform(df_test_labels)

print("Original training labels:")
print(df_train_labels.head(), "\n")

print("Encoded training labels (first 10):")
print(df_train_labels_prepared[:10])
```

```python
# Creating a preprocessing pipeline for numerical features
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


num_attribs = [ 'radius_mean', 'texture_mean', 'perimeter_mean',
                'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
                'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
                'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
                'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
                'fractal_dimension_se', 'radius_worst', 'texture_worst',
                'perimeter_worst', 'area_worst', 'smoothness_worst',
                'compactness_worst', 'concavity_worst', 'concave points_worst',
                'symmetry_worst', 'fractal_dimension_worst']

# Our data set has no missing values, but in case they appear
# We replace missing values with the median (SimpleImputer)
# We standardize features to zero mean and unit variance (StandardScaler)

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

preprocessing = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
    ])


# Applying pipeline
df_train_prepared = preprocessing.fit_transform(df_train)
df_test_prepared = preprocessing.transform(df_test)

print("Training data shape after preprocessing:", df_train_prepared.shape)
print("Test data shape after preprocessing:", df_test_prepared.shape)
```

```python
# Train SGDClassifier
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(df_train_prepared, df_train_labels_prepared)

print("SGDClassifier training completed successfully.")
```

```python
# Metrics: confusion matrix, precision, recall, F1-score, acccuracy
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score

# Make predictions on the training set
df_prediction_sgd = sgd_clf.predict(df_train_prepared)

# We compute evaluation metrics on training data
train_cm_sgd = confusion_matrix(df_train_labels_prepared, df_prediction_sgd)
train_precision_sgd = precision_score(df_train_labels_prepared, df_prediction_sgd)
train_recall_sgd = recall_score(df_train_labels_prepared, df_prediction_sgd)
train_sgd_f1 = f1_score(df_train_labels_prepared, df_prediction_sgd)
train_accuracy_sgd = accuracy_score(df_train_labels_prepared, df_prediction_sgd)


print("Results of evaluating the SGDClassifier on the TRAINING dataset:")
print("Confusion Matrix:\n", train_cm_sgd)
print(f"The precision of the model on the training dataset is: {train_precision_sgd:.2f}")
print(f"The recall of the model on the training dataset is: {train_recall_sgd:.2f}")
print(f"The F1-score of the model on the training dataset is: {train_sgd_f1:.2f}")
print(f"The accuracy of the model on the training dataset is: {train_accuracy_sgd:.2f}")
```

```python
# Evaluate the performance of the SGDClassifier on the TEST dataset

df_prediction_sgd_test = sgd_clf.predict(df_test_prepared)

test_cm_sgd = confusion_matrix(df_test_labels_prepared, df_prediction_sgd_test)
test_precision_sgd = precision_score(df_test_labels_prepared, df_prediction_sgd_test)
test_recall_sgd = recall_score(df_test_labels_prepared, df_prediction_sgd_test)
test_sgd_f1 = f1_score(df_test_labels_prepared, df_prediction_sgd_test)
test_accuracy_sgd = accuracy_score(df_test_labels_prepared, df_prediction_sgd_test)

print("Results of evaluating the SGDClassifier on the TEST dataset:")
print("Confusion Matrix:\n", test_cm_sgd)
print(f"The precision of the model on the test dataset is: {test_precision_sgd:.2f}")
print(f"The recall of the model on the test dataset is: {test_recall_sgd:.2f}")
print(f"The F1-score of the model on the test dataset is: {test_sgd_f1:.2f}")
print(f"The accuracy of the model on the test dataset is: {test_accuracy_sgd:.2f}")
```

```python
# Train a Linear Support Vector Machine (SVM) classifier
from sklearn.svm import LinearSVC

svm_clf = LinearSVC(C=2, random_state=42)
svm_clf.fit(df_train_prepared, df_train_labels_prepared)

df_train_pred_clf= svm_clf.predict(df_train_prepared)
df_test_pred_clf = svm_clf.predict(df_test_prepared)

# Evaluate the Linear SVM model on the TRAINING dataset
cm_train_svm = confusion_matrix(df_train_labels_prepared, df_train_pred_clf)
precision_train_svm = precision_score(df_train_labels_prepared, df_train_pred_clf)
recall_train_svm = recall_score(df_train_labels_prepared, df_train_pred_clf)
f1_train_svm = f1_score(df_train_labels_prepared, df_train_pred_clf)
accuracy_train_svm = accuracy_score(df_train_labels_prepared, df_train_pred_clf)

print("Results of evaluating the Linear SVM model on the TRAINING dataset:")
print("Confusion Matrix:\n", cm_train_svm)
print(f"The precision of the Linear SVM model on the training dataset is: {precision_train_svm:.2f}")
print(f"The recall of the Linear SVM model on the training dataset is: {recall_train_svm:.2f}")
print(f"The F1-score of the Linear SVM model on the training dataset is: {f1_train_svm:.2f}")
print(f"The accuracy of the Linear SVM model on the training dataset is: {accuracy_train_svm:.2f}")


# Evaluate the Linear SVM model on the TEST dataset
cm_test_svm = confusion_matrix(df_test_labels_prepared, df_test_pred_clf)
precision_test_svm = precision_score(df_test_labels_prepared, df_test_pred_clf)
recall_test_svm = recall_score(df_test_labels_prepared, df_test_pred_clf)
f1_test_svm = f1_score(df_test_labels_prepared, df_test_pred_clf)
accuracy_test_svm = accuracy_score(df_test_labels_prepared, df_test_pred_clf)

print("\nResults of evaluating the Linear SVM model on the TEST dataset:")
print("Confusion Matrix:\n", cm_test_svm)
print(f"The precision of the Linear SVM model on the test dataset is: {precision_test_svm:.2f}")
print(f"The recall of the Linear SVM model on the test dataset is: {recall_test_svm:.2f}")
print(f"The F1-score of the Linear SVM model on the test dataset is: {f1_test_svm:.2f}")
print(f"The accuracy of the Linear SVM model on the test dataset is: {accuracy_test_svm:.2f}")
```

```python
# SVM with Polynomial Kernel (degree = 2)
from sklearn.model_selection import cross_val_predict
from sklearn.svm import SVC

# y values (labels): 0 = benign (B), 1 = malignant (M)
y_train = df_train_labels_prepared
y_test = df_test_labels_prepared

# Create SVM classifier with polynomial kernel of degree 2
poly_svm_clf = SVC( kernel="poly", degree=2, coef0=1, C=1, gamma="scale", random_state=42)

# Perform 5-fold cross-validation on the training dataset
y_train_pred_cv_poly = cross_val_predict(poly_svm_clf, df_train_prepared, y_train, cv=5)
cm_cv_poly = confusion_matrix(y_train, y_train_pred_cv_poly)

print("\nSVM Poly (degree=2) – Confusion Matrix using cross_val_predict():")
print(cm_cv_poly)
print()

# Train the model on the prepared train data
poly_svm_clf.fit(df_train_prepared, y_train)

# Predict on train and test data
y_train_pred_poly = poly_svm_clf.predict(df_train_prepared)
y_test_pred_poly = poly_svm_clf.predict(df_test_prepared)

# Compute performance metrics on the training dataset
train_cm_poly = confusion_matrix(y_train, y_train_pred_poly)
train_precision_poly = precision_score(y_train, y_train_pred_poly)
train_recall_poly = recall_score(y_train, y_train_pred_poly)
train_f1_poly = f1_score(y_train, y_train_pred_poly)
train_accuracy_poly = accuracy_score(y_train, y_train_pred_poly)

print("Results of evaluating the Polynomial SVM (degree=2) model on the TRAINING dataset:")
print("Confusion Matrix:\n", train_cm_poly)
print(f"The precision of the Polynomial SVM model on the training dataset is: {train_precision_poly:.2f}")
print(f"The recall of the Polynomial SVM model on the training dataset is: {train_recall_poly:.2f}")
print(f"The F1-score of the Polynomial SVM model on the training dataset is: {train_f1_poly:.2f}")
print(f"The accuracy of the Polynomial SVM model on the training dataset is: {train_accuracy_poly:.2f}")
print()

# Compute performance metrics on the test dataset
test_cm_poly = confusion_matrix(y_test, y_test_pred_poly)
test_precision_poly = precision_score(y_test, y_test_pred_poly)
test_recall_poly = recall_score(y_test, y_test_pred_poly)
test_f1_poly = f1_score(y_test, y_test_pred_poly)
test_accuracy_poly = accuracy_score(y_test, y_test_pred_poly)

print("Results of evaluating the Polynomial SVM (degree=2) model on the TEST dataset:")
print("Confusion Matrix:\n", test_cm_poly)
print(f"The precision of the Polynomial SVM model on the test dataset is: {test_precision_poly:.2f}")
print(f"The recall of the Polynomial SVM model on the test dataset is: {test_recall_poly:.2f}")
print(f"The F1-score of the Polynomial SVM model on the test dataset is: {test_f1_poly:.2f}")
print(f"The accuracy of the Polynomial SVM model on the test dataset is: {test_accuracy_poly:.2f}")
```
